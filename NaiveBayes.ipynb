```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB
from sklearn.inspection import DecisionBoundaryDisplay
import sklearn.metrics as sm
data_set=pd.read_csv("apple_quality1.csv")
x=data_set.iloc[:,[2,3]].values
y=data_set.iloc[:,8].values
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.07,random_state=0)
std=StandardScaler()
x_train=std.fit_transform(x_train)
x_test=std.transform(x_test)
classifier=GaussianNB()
classifier.fit(x_train,y_train)
y_pred=classifier.predict(x_test)
print(y_pred)
```

```python
print(sm.confusion_matrix(y_test,y_pred))
print(sm.accuracy_score(y_test,y_pred))
print(sm.precision_score(y_test,y_pred,average="weighted"))
```
