```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
data_set=pd.read_csv("apple_quality1.csv")
x=data_set.iloc[:,[1,3]].values
points = []
for i in range(1,8):
    model=KMeans(n_clusters=i,init="k-means++",random_state=0,n_init=10)
    model.fit(x)
    points.append(model.inertia_)
plt.title("elbow method graph")
plt.plot(range(1,8),points)
plt.xlabel(data_set.columns[1])
plt.ylabel(data_set.columns[3])

plt.show()
```

```python
no_of_cluster=4
model=KMeans(no_of_cluster,init="k-means++",random_state=0,n_init=10)
pred=model.fit_predict(x)
for i in range(no_of_cluster):
    rgb=(np.random.rand(),np.random.rand(),np.random.rand())
    plt.scatter(x[pred==i,0],x[pred==i,1],color=[rgb],label=f"Cluster{i+1}")
plt.title("KMeans Clustering")
plt.xlabel(data_set.columns[1])
plt.ylabel(data_set.columns[3])
plt.legend()
plt.show()
```
