```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch

dataset = pd.read_csv("apple_quality1.csv")

X = dataset.iloc[:,[6,7]].values

dendo = sch.dendrogram(sch.linkage(X,method="ward")) # linkage takes the dataset and the linkage method calculates adn returns the hierarchical cluster and dendrogram function takes it can draw the dendrogram

# some pther linkage methods are single ,complete etc here we used ward which minimizes the varience between the clusters

plt.title("Dendrogram")
plt.xlabel(dataset.columns[6])
plt.ylabel(dataset. columns[7])
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering as AC

no_of_cluster = 5

model = AC(n_clusters = no_of_cluster,metric="euclidean",linkage="ward")
prediction = model.fit_predict(X)

for i in range(no_of_cluster):
    rgb = (np.random.rand(),np.random.rand(),np.random.rand())
    plt.scatter(X[prediction==i,0],X[prediction==i,1],color=[rgb],label=f"Cluster {i + 1}")

plt.title("Hierarchical Clustering")
plt.xlabel(dataset.columns[6])
plt.ylabel(dataset.columns[7])
plt.legend()
plt.show()
```
