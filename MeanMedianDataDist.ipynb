```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statistics as stat
data_set = np.random.uniform(1,175,10000)

data = pd.read_csv("pokemon.csv")
X=data.iloc[:,[3]].values
Y=data.iloc[:,[7]].values
print(np.mean(data_set))
print(np.median(data_set))
print(np.std(data_set))
print(stat.mode(data_set))
plt.hist(data_set,1000,label="Histogram")
plt.scatter(X,Y,c="red",label="Scatter Plot")
plt.title("Data Distribution")
plt.xlabel(data.columns[3])
plt.ylabel(data.columns[7])
plt.legend()
plt.show()
```
